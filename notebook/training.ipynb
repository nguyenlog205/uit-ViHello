```python
import pandas as pd
```

```python
df = pd.read_json('../checkpoints/history.json')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
loss_at_200 = df.loc[df['epoch'] == 200, 'loss'].values[0]

plt.figure(figsize=(11, 3.5))
plt.title('Training loss')
plt.plot(df['epoch'], df['loss'], label='Training loss')

plt.axhline(y=loss_at_200, color='red', linestyle='--', label=f'Loss at epoch 200: {loss_at_200:.4f}')

plt.legend()
plt.savefig('../img/training_loss.png')
```
